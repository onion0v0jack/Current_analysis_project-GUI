```python
import time
import os
import pywt 
import configparser
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from sklearn.decomposition import PCA

plt.style.use('ggplot')   # 設定畫圖風格
plt.rcParams['font.sans-serif'] = ['SimHei'] # 設定相容中文 
plt.rcParams['axes.unicode_minus'] = False
pd.options.mode.chained_assignment = None
```

```python
def data_truncate(Input_data, column, ct, cut1, cut_length):
    #截斷對齊，前截斷並起算序列長度
    scan_data = Input_data.loc[:, column].values
    for index, value in enumerate(scan_data):
        if (index > 0) and (value > cut1) and (value - scan_data[index - 1] > 0) and (scan_data[index + 1] - value > 0):
            trun1 = index
            break
    Input_data = Input_data.loc[trun1 : (trun1 + cut_length - 1), :]
    Input_data.reset_index(drop = True, inplace = True)
    Input_data.loc[:, ct] = Input_data.loc[:, ct] - Input_data.loc[:, ct].min()
    return Input_data

#小波分解函數
def dwt_split(Data, column, waveletname):
    data = Data.loc[:, column].values
    (cA1, cD1) = pywt.dwt(data, waveletname)
    (cA2, cD2) = pywt.dwt(cA1, waveletname)
    (cA3, cD3) = pywt.dwt(cA2, waveletname)
    (cA4, cD4) = pywt.dwt(cA3, waveletname)
    return cD1, cD2, cD3, cD4, cA4
```

```python
main_addr = '0923_10號機'
config = configparser.ConfigParser()
config.read(f'程式執行區\config\Configure_for_demo.ini', encoding = "utf-8-sig")

# addr = config.get('Data', 'Dataset_addr') #No
column_show = config.get('Data', 'Column_show') #Current
column_time = config.get('Data', 'Column_time') #CatchTime
cts = float(config.get('Preprocessing', 'Current_trun_start')) #30
cl = int(config.get('Preprocessing', 'Cut_length')) #500
wavelet_name = config.get('Analysis', 'Wavelet') #db1
n_pc = int(config.get('Analysis', 'PC_number')) #2
# MSE_thres_List = [float(x.strip()) for x in config.get('Analysis', 'MSE_thres_list').split(',')] #0.0001, 0.0005, 0.0025, 0.005, 0.05 
```

```python
%matplotlib inline
fig, ax = plt.subplots(2, 2, figsize = (17, 5), subplot_kw = {'facecolor':'#EEEEEE'}, edgecolor = "white")
if len(os.listdir(f'data/{main_addr}/train')) > 0:
    for count, filename in enumerate(os.listdir(f'data/{main_addr}/train')):
        Data = pd.read_table(r'data/{}/train/{}'.format(main_addr, filename), sep = ',')
        ax[0, 0].plot(Data.loc[:, column_time], Data.loc[:, column_show], linewidth = 1, color = 'dimgrey', alpha = 0.4)
        Data = data_truncate(Input_data = Data, column = column_show, ct = column_time, cut1 = cts, cut_length = cl)
        ax[1, 0].plot(Data.loc[:, column_time], Data.loc[:, column_show], linewidth = 1, color = 'dimgrey', alpha = 0.4)
    ax[0, 0].set_title(f'{count + 1}組訓練資料的原區段曲線', fontproperties = FontProperties(fname = "SimHei.ttf", size = 14))
    ax[1, 0].set_title(f'{count + 1}組經截斷後的訓練資料', fontproperties = FontProperties(fname = "SimHei.ttf", size = 14))
else:
    ax[0, 0].set_title(f'0組訓練資料的原區段曲線', fontproperties = FontProperties(fname = "SimHei.ttf", size = 14))
    ax[1, 0].set_title(f'0組經截斷後的訓練資料', fontproperties = FontProperties(fname = "SimHei.ttf", size = 14))

pass_data_count = 0
for count, filename in enumerate(os.listdir(f'data/{main_addr}/test')):
    Data = pd.read_table(r'data/{}/test/{}'.format(main_addr, filename), sep = ',')
    ax[0, 1].plot(Data.loc[:, column_time], Data.loc[:, column_show], linewidth = 1, color = 'dimgrey', alpha = 0.4)
    if len(Data) > cl and count < 900 and Data.loc[:, column_show].values.max() > cts:
        try:
            Data = data_truncate(Input_data = Data, column = column_show, ct = column_time, cut1 = cts, cut_length = cl)
            ax[1, 1].plot(Data.loc[:, column_time], Data.loc[:, column_show], linewidth = 1, color = 'dimgrey', alpha = 0.4)
        except:
            continue
        pass_data_count += 1
    print(filename)
ax[0, 1].set_title(f'{count + 1}組測試資料的原區段曲線', fontproperties = FontProperties(fname = "SimHei.ttf", size = 14))
ax[1, 1].set_title(f'{pass_data_count}組經截斷後的測試資料', fontproperties = FontProperties(fname = "SimHei.ttf", size = 14))

for axi in range(2):
    for axj in range(2):
        ax[axi, axj].set_ylabel(column_show)
        ax[axi, axj].set_xlabel('Time(sec)')

plt.subplots_adjust(wspace = 0.1, hspace = 0.5)#調整子圖間距
plt.show()
```

```python
for count, filename in enumerate(os.listdir(f'{main_addr}/train_data')):
    #載入資料
    Data = pd.read_table(r'{}\train_data\{}'.format(main_addr, filename), sep = ',')
    Data = Data[[column_time, column_show]]
    #截斷對齊，前截斷並起算序列長度
    Data = data_truncate(Input_data = Data, column = column_show, ct = column_time, cut1 = cts, cut_length = cl)
    
    #將電流正規化
    X = Data.loc[:, column_show].values
    X_mean, X_std = X.mean(), X.std()
    Data.loc[:, 'Normalized_Current'] = (X - X_mean)/X_std
    
    #小波分解
    cD1, cD2, cD3, cD4, cA4 = dwt_split(Data, column = 'Normalized_Current', waveletname = wavelet_name)
    
    #講每個趨勢的分解小波個別合併
    if count == 0:
        D1_bind, D2_bind, D3_bind, D4_bind, A4_bind = cD1, cD2, cD3, cD4, cA4
    else:
        D1_bind = np.column_stack((D1_bind, cD1))
        D2_bind = np.column_stack((D2_bind, cD2))
        D3_bind = np.column_stack((D3_bind, cD3))
        D4_bind = np.column_stack((D4_bind, cD4))
        A4_bind = np.column_stack((A4_bind, cA4))
            
#對各分解小波訓練各自的PCA模型，注意餵進去的資料是訓練集
pca_D1, pca_D2, pca_D3, pca_D4, pca_A4 = PCA(n_components = n_pc), PCA(n_components = n_pc), PCA(n_components = n_pc), PCA(n_components = n_pc), PCA(n_components = n_pc)
pca_D1.fit(D1_bind)
pca_D2.fit(D2_bind)
pca_D3.fit(D3_bind)
pca_D4.fit(D4_bind)
pca_A4.fit(A4_bind)
print('各通道PCA模型完成。\n')

bind_data_list = [D1_bind, D2_bind, D3_bind, D4_bind, A4_bind]
pca_model_list = [pca_D1, pca_D2, pca_D3, pca_D4, pca_A4]
ylabel_list = ['D1', 'D2', 'D3', 'D4', 'A4']

print('各主成分解釋變量累積比例:')
for point in range(len(pca_model_list)):
    print('pca_{}: {}'.format(ylabel_list[point], pca_model_list[point].explained_variance_ratio_.cumsum()))
print('*'*50)

cum_val_pass = sum([pca_model.explained_variance_ratio_.cumsum()[-1] > 0.9 for pca_model in pca_model_list])
if cum_val_pass < 5: print(f'目前主成分個數:{n_pc} 尚有通道的累積解釋變異不足90%，建議增加主成分個數\n')
else: print('各通道的累積解釋變異超過0.9，可做後續分析\n')
```

```python
%matplotlib inline
fig, ax = plt.subplots(5, 1, figsize = (15, 7), subplot_kw = {'facecolor':'#EEEEEE'}, edgecolor = "white", sharex = True)
bdl, pcal = bind_data_list, pca_model_list
tfl = os.listdir(f'{main_addr}/train_data')
MESthresl = MSE_thres_List ###
new_MESthresl_1, new_MESthresl_2, new_MESthresl_3 = [], [], []
ylabell = ylabel_list

for point in range(len(bdl)):
    MSE_series = np.zeros(len(tfl))
    Real = bdl[point][:, 0:len(tfl)]
    Pred = pcal[point].inverse_transform(pcal[point].fit_transform(Real))
    for iterate in range(len(tfl)):
        MSE_series[iterate] = np.mean((Real[:, iterate] - Pred[:, iterate])**2, axis = 0)  #### 
#     MSE_sorted = np.sort(MSE_series)
#     target_MSE_1 = np.diff(MSE_sorted).argsort()[-1]
#     target_MSE_2 = np.diff(MSE_sorted).argsort()[-2]
#     new_MESthresl_1.append(np.mean([MSE_sorted[target_MSE_1 + 1], MSE_sorted[target_MSE_1]]))
#     new_MESthresl_2.append(np.mean([MSE_sorted[target_MSE_2 + 1], MSE_sorted[target_MSE_2]]))
    new_MESthresl_3.append(np.mean(MSE_series) + np.std(MSE_series))
    
    ax[point].plot(range(len(tfl)), MSE_series, linewidth = 1, color = 'dimgrey', alpha = 0.6)
    ax[point].axhline(y = MESthresl[point], color = 'k')  #############
    #ax[point].axhline(y = new_MESthresl_1[point], color = 'r')  #############
    #ax[point].axhline(y = new_MESthresl_2[point], color = 'b')  #############
    ax[point].axhline(y = new_MESthresl_3[point], color = 'g')  #############
    ax[point].set_ylabel(ylabell[point], rotation = 0, labelpad = 20)

ax[0].set_title(f'訓練資料預測MSE', fontproperties = FontProperties(fname = "SimHei.ttf", size = 14))
ax[len(bdl) - 1].set_xlabel('訓練資料模次', fontproperties = FontProperties(fname = "SimHei.ttf", size = 14))

plt.subplots_adjust(wspace = 0.25)#調整子圖間距
plt.show()
```
